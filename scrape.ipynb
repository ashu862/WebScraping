```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_flipkart_health_products():
    base_url = 'https://www.flipkart.com/home-improvement/home-utilities-and-organizers/laundry/ironing-mats/pr?sid=h1m,iee,3a7,nlk&otracker=categorytree={}'
    
    titles = []
    max_prices = []
    discounted_prices = []
    num_reviews = []
    ratings = []
    descriptions = []
    urls = []

    for page_number in range(1, 15):  # Scraping data from first 100 pages
        url = base_url.format(page_number)
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"Failed to fetch page {page_number}. URL: {url}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')
        products = soup.find_all('div', class_='_1AtVbE')

        for product in products:
            title = product.find('a', class_='s1Q9rs')
            if title:
                title_text = title.text.strip()
                if title_text not in titles:  # Checking for unique title products
                    titles.append(title_text)
                else:
                    continue  # Skip if the title already exists
            else:
                print(f"Title not found for a product on page {page_number}. URL: {url}")
                continue  # Skip if title not found

            max_price = product.find('div', class_='_30jeq3')
            if max_price:
                max_prices.append(max_price.text.strip())
            else:
                print(f"Maximum price not found for a product on page {page_number}. URL: {url}")
                max_prices.append('')

            discounted_price = product.find('div', class_='_3I9_wc')
            if discounted_price:
                discounted_prices.append(discounted_price.text.strip())
            else:
                print(f"Discounted price not found for a product on page {page_number}. URL: {url}")
                discounted_prices.append('')

            num_review = product.find('span', class_='_2_R_DZ')
            if num_review:
                num_reviews.append(num_review.text.strip())
            else:
                print(f"Number of reviews not found for a product on page {page_number}. URL: {url}")
                num_reviews.append('')

            rating = product.find('div', class_='gUuXy-')
            if rating:
                ratings.append(rating.text.strip())
            else:
                print(f"Rating not found for a product on page {page_number}. URL: {url}")
                ratings.append('')

            product_url = product.find('a', class_='s1Q9rs')
            if product_url:
                product_detail_url = "https://www.flipkart.com" + product_url.get('href')
                description = get_product_description(product_detail_url)
                descriptions.append(description)
                urls.append(product_detail_url)
            else:
                print(f"URL not found for a product on page {page_number}. URL: {url}")
                descriptions.append('')
                urls.append('')

    df = pd.DataFrame({
        'Title': titles,
        'Maximum Price': max_prices,
        'Discounted Price': discounted_prices,
        'Number of Reviews': num_reviews,
        'Rating': ratings,
        'Description': descriptions,
        'URL': urls
    })

    return df

# Rest of the code remains the same
def get_product_description(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        description = soup.find('div', {'class': '_1mXcCf'})
        if description:
            return description.text.strip()
        else:
            print(f"Description not found for URL: {url}")
    else:
        print(f"Failed to fetch URL: {url}. Status code: {response.status_code}")
    return ''

def save_to_csv(dataframe, filename):
    dataframe.to_csv(filename, index=False)
    print(f'Data has been successfully saved to {filename}')

if __name__ == "__main__":
    health_products_data = scrape_flipkart_health_products()
    if not health_products_data.empty:
        save_to_csv(health_products_data, 'tool_32.csv')
    else:
        print("No data scraped. Check the script and website structure.")
```
